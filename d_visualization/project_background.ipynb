```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import rc
rc("font", family = "AppleGothic")
```

```python
stat = pd.read_csv("./datasets/year_age_stats.csv")
stat.head()

```

```python
stat.info()
```

```python
stat = stat[["Year","Total"]]
```

```python
data = {"Year": ["2022","2023","2024"],
       "Total": [6554031 ,22715841 ,28686435 ]}


data  = pd.DataFrame(data)
data
```

```python
stat = pd.concat([stat, pd.DataFrame(data)], ignore_index=True)
```

```python
years = stat.Year.tolist() # 인덱스 리스트 넘기기

x_pos = np.arange(len(years))
hp = stat.Total.tolist() 

bar_list = plt.bar(x_pos,hp,alpha = 0.5)

# 마지막 바 만 빨간색으로 변경
bar_list[-1].set_color("r") # .set(color= "r") 이것도 가능

# 아래와 같은 방법도 있음
# plt.hlines(alc_mean,0,6,color="black",linestyle="solid") 
plt.xticks(x_pos,years,rotation=90)
plt.title("내국인 해외 관광객 수")
plt.show()
```
